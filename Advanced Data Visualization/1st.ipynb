```python
x=np.arange(-10,11,1)
y=x**2

plt.plot(x,y,color="green",marker='o')
plt.xlabel('my x-values')
plt.ylabel('my y-values')
plt.title("Squares function")
plt.show()
#Marker puts desired things as if marker="o" then it will put block ther

#there is alos l=something like linestyle
```

```python
x=np.arange(-10,11,1)
y=x**2

plt.plot(x,y,color="green",marker='o',linestyle='dashed')
plt.xlabel('my x-values')
plt.ylabel('my y-values')
plt.title("Squares function")
plt.show()
```

```python
#Sometimes we do all this in very intyitive ways such as
plt.plot(x,y,'go--')
plt.xlabel('my x-values')
plt.ylabel('my y-values')
plt.title("Squares function")
plt.show()
#as go-- g for green o for marker and -- for dashed line 
```

```python
#Earler it was best fit in graph
plt.plot(x,y,'go--')
plt.xlabel('my x-values')
plt.ylabel('my y-values')
plt.title("Squares function")
plt.xlim(-11,20)
plt.show()
```

```python
x=np.arange(1,10,0.1)
y1=np.sin(x)
y2=np.tan(x)
y3=np.log(x)
```

```python
plt.subplot(2,3,1)
plt.plot(x,y1)

plt.subplot(2,3,3)
plt.plot(x,y2)

plt.subplot(2,3,5)
plt.plot(x,y3)

plt.show()
```

```python
fig=plt.figure()
```

```python
ax=fig.add_axes([0,1,1,1])
```

```python
ax.plot(x,y1)
```

```python
ax=fig.add_subplot([0,1,1,1])
```

## We did functional Matplotlib Programming
## Object Oriented Programming

```python
fig=plt.figure()
ax231=fig.add_subplot(2,3,1)
ax231.plot(x,y1)

ax233=fig.add_subplot(2,3,3)
ax233.plot(x,y2)
ax233.set_xlim(-5,10)

ax234=fig.add_subplot(2,3,4)
ax234.plot(x,y3)

plt.show()
```
