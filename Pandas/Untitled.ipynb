```python
wheather=pd.read_csv('weather_data.csv')
wheather.head()
```

```python
wheather.shape
```

```python
#I want the Location , Date_Time , and Temperature_C ko kuch nhi kare
pd.melt(wheather,id_vars=['Location','Date_Time','Temperature_C'])
#It takes all the column and convert it into the melt variable
#id_vars=JiskoKuch Nhi karna chahte
#var_name=varable ke jagah wahi ho jayega
#value_name=jo chahe wo hojayega as per need
```

```python
wheather_melt=pd.melt(wheather,id_vars=['Location','Date_Time','Temperature_C'],
                      var_name='day',
                      value_name='temp')
wheather_melt.head()
#var_name & value_name are two way to rename the column created by the melt 
#Wide form of Data is Rearranged in VAertical form as rows and columns
```

```python
wheather_melt.shape
#pd.melt() is like more of do not touch this and the rest will be melted into one 
```

```python
#Now suppose we have [year , month , element , day , temp]
#Now we will merge both of the element nd the temp inorder to get 
#max_temp,min_temp
```

```python
#If we have List No of days it will convrt it into NaN
```

```python
wheather_tidy=pd.pivot_table(wheather_melt,
                             index=['Location','Date_Time','day'],
                             columns=['temp'],
                             values='Temperature_C')
wheather_tidy
```
